````python
# Δημιουργία του κάθε κεφαλαίου σε ξεχωριστό αρχείο pdf, έτσι ώστε να δημιουργηθούν ερωταπαντήσεις από κάθε κεφάλαιο.

# import csv
# import os
# from PyPDF2 import PdfReader, PdfWriter

def split_pdf_by_chapter(pdf_path, csv_path, output_dir):
    # Δημιουργούμε τον φάκελο αν δεν υπάρχει
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Ανοίγουμε το αρχείο PDF για ανάγνωση
    reader = PdfReader(pdf_path)

    # Συνολικός αριθμός σελίδων
    total_pages = len(reader.pages)

    # Διαβάζουμε το CSV αρχείο για τα κεφάλαια και τις σελίδες
    chapters = []
    with open(csv_path, mode='r', encoding='utf-8') as csvfile:
        csv_reader = csv.DictReader(csvfile)
        for row in csv_reader:
            title = row['title']
            starting_page = int(row['starting_page']) #- 1  # Μετατροπή σε μηδενική βάση
            chapters.append({'title': title, 'starting_page': starting_page})

    # Υπολογίζουμε την ending_page για κάθε κεφάλαιο
    for i in range(len(chapters)):
        if i < len(chapters) - 1:
            chapters[i]['ending_page'] = chapters[i + 1]['starting_page']  - 1
        else:
            chapters[i]['ending_page'] = total_pages - 1  # Τελευταία σελίδα του PDF

    i = 1
    # Δημιουργούμε ένα ξεχωριστό αρχείο PDF για κάθε κεφάλαιο
    for chapter in chapters:
        writer = PdfWriter()
        title = chapter['title']
        start = chapter['starting_page']
        end = chapter['ending_page']

        # Επαλήθευση ορίων σελίδων
        if start < 0 or end >= total_pages or start > end:
            print(f"Σφάλμα: Οι σελίδες του κεφαλαίου '{title}' είναι εκτός ορίων. Αγνοείται.")
            continue

        # Προσθέτουμε τις σελίδες του κεφαλαίου στον writer
        for page_num in range(start, end + 1):
            writer.add_page(reader.pages[page_num])

        # Καθαρίζουμε το όνομα αρχείου από μη αποδεκτούς χαρακτήρες
        safe_title = "".join(c if c.isalnum() or c.isspace() else "_" for c in title).strip()
        # Προσθέτουμε το "i" στον τίτλο αν δεν περιέχει τη λέξη "Ενότητα"
        if "Ενότητα" not in title:
            safe_title = "κεφάλαιο " + str(i) + " " + safe_title
            i += 1
        output_filename = f"{output_dir}/{safe_title}.pdf"

        # Αποθηκεύουμε το νέο PDF αρχείο
        with open(output_filename, "wb") as output_pdf:
            writer.write(output_pdf)

        print(f"Αποθηκεύτηκε το κεφάλαιο '{title}' στο '{output_filename}'.")

pdf_path = "/content/drive/MyDrive/Διπλωματική/Books/Geography_E_Class/10-0239-01_V2_Geografia_E-Dimotikou_Vivlio-Mathiti.pdf"  # Διαδρομή του αρχικού PDF
csv_path = "/content/drive/MyDrive/Διπλωματική/Books/Geography_E_Class/Csv/chapters_output.csv"  # Διαδρομή του CSV αρχείου
output_dir = "/content/drive/MyDrive/Διπλωματική/Books/Geography_E_Class/Chapters"  # Φάκελος για τα αρχεία κεφαλαίων
split_pdf_by_chapter(pdf_path, csv_path, output_dir)
````

````python
# Εξαγωγή εικόνων από αρχείο PDF και αποθήκευση σε φακέλους ανά κεφάλαιο

!pip install PyMuPDF Pillow
!pip install PyPDF2

import os
from PyPDF2 import PdfReader
from PIL import Image
import fitz

def extract_images_from_pdf(pdf_path, output_folder):
    """
    Extracts all images from a PDF file and saves them to a specified folder.
    """
    # Άνοιγμα του PDF
    pdf_document = fitz.open(pdf_path)

    # Loop πρόσβασης σε όλες τις σελίδες
    for page_num in range(len(pdf_document)):
        page = pdf_document[page_num]
        images = page.get_images(full=True)

        for img_index, img in enumerate(images):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]
            image_ext = base_image["ext"]
            image_name = f"page_{page_num + 1}_img_{img_index + 1}.{image_ext}"

            # Αποθήκευση εικόνας στον φάκελο εξόδου
            with open(os.path.join(output_folder, image_name), "wb") as img_file:
                img_file.write(image_bytes)

def process_all_pdfs(input_folder, output_base_folder):
    """
    Processes all PDF files in a folder, extracts images, and saves them
    to chapter-specific folders.
    """
    if not os.path.exists(output_base_folder):
        os.makedirs(output_base_folder)

    for pdf_file in os.listdir(input_folder):
        if pdf_file.endswith('.pdf'):
            chapter_name = os.path.splitext(pdf_file)[0]
            chapter_folder = os.path.join(output_base_folder, chapter_name)
            os.makedirs(chapter_folder, exist_ok=True)

            pdf_path = os.path.join(input_folder, pdf_file)
            extract_images_from_pdf(pdf_path, chapter_folder)
            print(f"Images from '{pdf_file}' saved to '{chapter_folder}'.")

# Φάκελος εισόδου που υπάρχουν τα αρχεία PDF ανά κεφάλαιο και φάκελος εξόδου των φακέλων των εικόνων ανά κεφάλαιο
input_pdf_folder = "/content/drive/MyDrive/Διπλωματική/Books/Geography_E_Class/Chapters"
output_folder = "/content/drive/MyDrive/Διπλωματική/Books/Geography_E_Class/Images_per_chapter"

process_all_pdfs(input_pdf_folder, output_folder)
````

````python
# 29-12-2024 Δίνει το όνομα του αρχείου της εικόνας, την περιγραφή της και το path από κάθε υποφάκελο που έχει τις εικόνες
# του κάθε κεφαλαίου αντίστοιχα. Αποθηκεύει τα δεδομένα σε ένα txt αρχείο σε κάθε υποφάκελο. Tο πρόβλημα με την εξάντληση των ορίων
# των API λύθηκε αλλάζοντας μοντέλο σε gemini 1.5 flash και χρονική καθυστέρηση 10 δευτερολέπτων.

import os
import google.generativeai as genai
import json
import time

# API keys setup. Αντικατάσταση ...... με τα keys
api_keys = [
    '......',
    '......',
    '......',
    '......',
    '......'
]
current_key_index = 0
last_request_time = 0
genai.configure(api_key=api_keys[current_key_index])

def switch_api_key():
    """
    Switch to the next API key in the list and reconfigure the GenAI client.
    """
    global current_key_index, last_request_time
    current_key_index = (current_key_index + 1) % len(api_keys)
    genai.configure(api_key=api_keys[current_key_index])
    last_request_time = time.time()  # Επαναφορά χρόνου τελευταίου αιτήματος
    print(f"Switched to API key index: {current_key_index}")

def enforce_request_delay():
    """
    Ensure a minimum delay of 2 seconds between requests for the current API key.
    """
    global last_request_time
    elapsed_time = time.time() - last_request_time
    if elapsed_time < 10:
        time.sleep(10 - elapsed_time)
    last_request_time = time.time()

def upload_the_file_to_gemini(path, mime_type="image/jpeg"):
    """
    Upload an image file to Gemini with API rate-limiting.
    """
    while True:
        enforce_request_delay()
        try:
            file = genai.upload_file(path, mime_type=mime_type)
            return file
        except Exception as e:
            print(f"Error with API key {current_key_index}: {e}")
            switch_api_key()

def process_images_summarize_and_paths(parent_folder):
    """
    Process images in subfolders, generate summaries, and match them with URLs.
    Args:
        parent_folder: Path to the parent folder containing subfolders.
    """
    # Prompt για περίληψη εικόνας
    prompt = """
    You are an assistant tasked with summarizing images for retrieval. \
    These summaries will be embedded and used to retrieve the raw image. \
    Give a concise summary of the image that is well optimized for retrieval."""
    for subfolder in os.listdir(parent_folder):
        subfolder_path = os.path.join(parent_folder, subfolder)
        # Παράλειψη ότι δεν είναι path φακέλων
        if not os.path.isdir(subfolder_path):
            continue
        # Δημιουργία αρχείο κειμένου στον υποφάκελο
        output_txt_file = os.path.join(subfolder_path, f"{subfolder}_image_summaries.txt")
        print(f"Processing subfolder: {subfolder_path}")
        results = []
        with open(output_txt_file, "w") as txt_file:
            for img_file in sorted(os.listdir(subfolder_path)):
                if img_file.endswith((".jpeg", ".jpg", ".png")):
                    img_path = os.path.join(subfolder_path, img_file)

                    # Φόρτωση εικόνας στο Gemini
                    try:
                        gemini_file = upload_the_file_to_gemini(img_path)
                        gemini_url = gemini_file.uri
                    except Exception as e:
                        print(f"Failed to upload {img_file} to Gemini: {e}")
                        continue

                    # Δημιουγία περίληψης
                    while True:
                        enforce_request_delay()
                        try:
                            model = genai.GenerativeModel("gemini-1.5-flash")
                            response = model.generate_content([prompt, gemini_file])
                            response_text = response.text.strip("```json").strip("```")
                            break
                        except Exception as e:
                            print(f"Error with API key {current_key_index}: {e}")
                            switch_api_key()

                    # Προσθήκη αποτελέσματος
                    result = {
                        "file_name": img_file,
                        "summary": response_text,
                        "path": img_path,
                    }
                    results.append(result)

                    # Εγγραφή σε αρχείο TXT
                    txt_file.write(
                        f"Filename: {img_file}\n"
                        f"Summary: {response_text}\n"
                        f"Path: {img_path}\n\n"
                    )

                    # Καταγραφή προόδου
                    print(f"Processed: {img_file}")
                    print(f"Summary: {response_text}")
                    print(f"Path: {img_path}")

        # Αποθήκευση αποτελεσμάτων σε αρχείο JSON στον υποφάκελο
        output_json_file = os.path.join(subfolder_path, f"{subfolder}_image_summaries.json")
        with open(output_json_file, "w") as json_file:
            json.dump(results, json_file, ensure_ascii=False, indent=4)

        print(f"Results saved to {output_json_file}")
        print(f"Log saved to {output_txt_file}")

# Είσοδοι
parent_folder = "/content/drive/MyDrive/Διπλωματική/Books/Geography_E_Class/Images_per_chapter"

# Επεξεργασία όλων των υποφακέλων
process_images_summarize_and_paths(parent_folder)
````

````python
# Φορτώνει το txt αρχείο από to drive που περιέχει τις περιγραφές στα αγγλικά.

def load_image_summaries(file_path):
    """
    Load image summaries from a text file and store them as tuples.

    Args:
        file_path (str): Path to the text file containing the image summaries.

    Returns:
        list of tuples: Each tuple contains (Filename, Summary, Path).
    """
    with open(file_path, "r", encoding="utf-8") as file:
        lines = file.readlines()

    results = []
    current_filename = None
    current_summary = None
    current_path = None

    for line in lines:
        line = line.strip()  # Αφαίρεση leading/trailing whitespace

        if line.startswith("Filename:"):
            current_filename = line.split(":", 1)[1].strip()
        elif line.startswith("Summary:"):
            current_summary = line.split(":", 1)[1].strip()
        elif line.startswith("Path:"):
            current_path = line.split(":", 1)[1].strip()
            # Προσθήκη πλειάδας όταν όλα τα πεδια συμπληρωθούν
            if current_filename and current_summary and current_path:
                results.append((current_filename, current_summary, current_path))
                current_filename = current_summary = current_path = None  # Επαναφορά επόμενης εισόδου

    return results

# Path αρχείου για φόρτωση
file_path = "/content/drive/MyDrive/Διπλωματική/Books/Geography_E_Class/all_txt_in_one.txt"
img_summaries = load_image_summaries(file_path)

# Εκτύπωση αποτελεσμάτων
for entry in img_summaries:
    print(entry)
````
